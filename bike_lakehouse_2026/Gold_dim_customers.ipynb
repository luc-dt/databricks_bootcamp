# Init

```python
import pyspark.sql.functions as F
from pyspark.sql.window import Window
```

# Business Transformation and Modeling

```python
query = """
SELECT 
    ROW_NUMBER() OVER (ORDER BY ci.customer_id) AS customer_key,
    ci.customer_id AS customer_id,
    ci.first_name,
    ci.last_name
FROM workspace.silver.cust_customers AS ci
"""
df = spark.sql(query)
```

# Write it to Gold Table

```python
(
    df.write
        .mode("overwrite")
        .format("delta")
        .saveAsTable("workspace.gold.dim_customers")
)
```
